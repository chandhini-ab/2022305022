```python
# Install and import everything
!pip install google-generativeai PyPDF2

import PyPDF2
import google.generativeai as genai
from google.colab import files

# Step 1: Upload the resume
uploaded = files.upload()
resume_filename = list(uploaded.keys())[0]

# Step 2: Extract text from PDF
def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        text = ''
        for page in reader.pages:
            content = page.extract_text()
            if content:
                text += content
    return text

resume_text = extract_text_from_pdf(resume_filename)

# Step 3: Authenticate with Gemini (get API key from https://makersuite.google.com/app/apikey)
genai.configure(api_key="AIzaSyDNFCd616Avp6JykfgMdkN22dGH7aKBADs")  # <-- Replace with your API key

# Step 4: Load Gemini model
model = genai.GenerativeModel('gemini-1.5-pro')

# Step 5: Prompt and generate feedback
prompt = f"""
You are an expert career coach and resume reviewer.

Please review the following resume and provide:
- Overall strengths and weaknesses
- Suggestions to improve formatting and content
- Specific areas to add or remove
- Tone, clarity, grammar feedback
- ATS (Applicant Tracking System) friendliness score out of 10

Resume:
\"\"\"
{resume_text}
\"\"\"
"""

response = model.generate_content(prompt)

# Step 6: Display feedback
print("üìÑ Gemini Resume Review:\n")
print(response.text)
```
